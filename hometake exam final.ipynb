```python
#Exercise 2

#step 1: retrieving the data about the waste collection system in Amsterdam from the Amsterdam API

#in order to acess APIs
import requests

#in order to handle json responses
import json
from pprint import pprint
from pandas import json_normalize

#for general data halding
import pandas as pd
```

```python
#Step 1: retriving the APIs
#Now, I am retrieving the first API
r = requests.get("https://api.data.amsterdam.nl/afval/v1/containers/")

data = r.json()
print(data.keys())
pprint(data)


#Now, I am retrieving the second API

r = requests.get("https://api.data.amsterdam.nl/afval/v1/wells/")
data1 = r.json()
print(data1.keys())
pprint(data1)

#Now, I am retrieving the thirs API

r = requests.get("https://api.data.amsterdam.nl/afval/v1/sites/")
data2 = r.json()
print(data2.keys())
pprint(data2)
```

```python
#retrieving multiple pages of the first API and nflattering the data to get it into a table
url_containers = "https://api.data.amsterdam.nl/afval/v1/containers/?page="

allitems = []
for i in range(1,200):
    r = requests.get(url_containers+f"{i}")
    data = r.json()
    if not "results" in data:
        print(f"Retrieved{len(allitems)}," "got all results")
        break
    allitems.extend(data["results"])
    i += 1

data_containers = json_normalize(allitems)
data_containers.head()
```

```python
#retrieving multiple pages of the second API and nflattering the data to get it into a table
url_containers1 = "https://api.data.amsterdam.nl/afval/v1/wells/?page="

allitems = []
for i in range(1,200):
    r = requests.get(url_containers1+f"{i}")
    data1 = r.json()
    if not "results" in data1:
        print(f"Retrieved{len(allitems)}," "got all results")
        break
    allitems.extend(data1["results"])
    i += 1

data_containers1 = json_normalize(allitems)
data_containers1.head()
```

```python
#retrieving multiple pages of the third API and nflattering the data to get it into a table
url_containers2 = "https://api.data.amsterdam.nl/afval/v1/sites/?page="

allitems = []
for i in range(1,150):
    r = requests.get(url_containers2+f"{i}")
    data2 = r.json()
    if not "results" in data2:
        print(f"Retrieved{len(allitems)}," "got all results")
        break
    allitems.extend(data2["results"])
    i += 1

data_containers2 = json_normalize(allitems)
data_containers2.head()
```

```python
#starting to clean up the first data table. First, I rename some variables so that it is easier to join the table later with the other tables. I then check that the variables names are indeed changed. 
data_containers_rename = data_containers.rename(columns={"id": "id_container",
                                                        "id_number": "id_number_container",
                                                       "well": "well_key",
                                                        "_links.self.href": "container.href",
                                                        "operational_date": "operational_date_container",
                                                        "warranty_date": "warranty_date_container",
                                                        "placing_date": "placing_date_container"})
data_containers_rename.head()

#I now drop the column dysplay. Indeed, as the id variable has been renamed as id_container, the _display variable is now redundant. 
data_containers_drop = data_containers_rename.drop(columns="_display")

#f


data_containers_drop['owner_identification'] = data_containers_drop['owner.name'].astype(str).str.cat(data_containers_drop['owner.id'].astype(str), sep=' ')
data_containers_drop.head()

data_containers_drop['waste_identification'] = data_containers_drop['waste_name'].astype(str).str.cat(data_containers_drop['waste_type'].astype(str), sep=' ')

data_containers_drop = data_containers_drop.drop(columns=["waste_type", "waste_name", "owner.id", "owner.name", "container_type.name"])
data_containers_drop.head()
```
